```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
birds=plt.imread("birds.jpeg")
plt.imshow(birds,cmap="gray")
```

```python
print(birds.size)
print(birds.dtype)
print(birds.ndim)
print(birds.shape)
print(birds)
```

```python
print("Number of rows=", birds.shape[0])
print("Number of columns=", birds.shape[1])
```

```python
bird_copy=birds[::]
plt.imshow(bird_copy)
```

```python
# Slicing the image into top half and using gray scale

part_1=int(bird_copy.shape[0]/2)
plt.imshow(bird_copy[0:part_1],cmap="gray")
```

```python
# Slicing the image into bottom half and using original scale
plt.imshow(bird_copy[part_1:])
```

```python
# Slicing the image vertically

part_2=int(bird_copy.shape[1]/2)
plt.imshow(bird_copy[:,0:part_2])
```

```python
plt.imshow(bird_copy[:,part_2:])
```

```python
# Shrinking the image or reducing size
plt.imshow(bird_copy[::8,::8])
```

```python
b_copy=birds[:]
plt.imshow(b_copy,cmap="gray")
```

```python
# Rotating the Image
b_copy[:50,:50]=255
b_copy[-50:,:50]=0
```

```python
plt.imshow(b_copy,cmap="gray")
```

```python
r1=plt.imshow(b_copy.T[::-1,:],cmap="gray") #Transpose the rows and column ad then reverse the rows
```
